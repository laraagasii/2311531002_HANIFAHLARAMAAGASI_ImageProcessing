```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
def konvolusi_5x5(input_path):
    # 1. Buka gambar
    img = Image.open(input_path).convert('RGB')
    img_array = np.array(img, dtype=float)
    tinggi, lebar, channel = img_array.shape
```

```python
# 2. Wadah hasil (Copy asli agar border aman)
input_path = '/content/image.jpg' # Asumsi path gambar, bisa diganti
img = Image.open(input_path).convert('RGB')
img_array = np.array(img, dtype=float)
hasil_array = np.copy(img_array)
offset = 2
```

```python
# 3. Proses Konvolusi per Channel
tinggi, lebar, channel = img_array.shape
for c in range(channel):
    for i in range(offset, tinggi - offset):
        for j in range(offset, lebar - offset):
            # Logika: Ambil area tetangga 5x5
            area = img_array[i-2 : i+3, j-2 : j+3, c]
            # Hitung rata-rata (Kernel Rataan 5x5)
            hasil_array[i, j, c] = np.sum(area) / 25
```

```python
# 4. Tampilkan Perbandingan
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(img)
ax[0].set_title("Citra Asli")
ax[1].imshow(hasil_array.astype(np.uint8))
ax[1].set_title("Hasil Konvolusi 5x5 (Blur)")
plt.show()
```
